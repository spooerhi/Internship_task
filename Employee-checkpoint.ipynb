### Bivariate Analysis

```python
df=data[['job_level','rating','awards','certifications','salary','satisfied','age','Years_of_ Experience']]
```

```python
# heatmap

sns.set(rc={'figure.figsize':(14,8)})
sns.heatmap(df.corr(),annot=True, fmt=".1f")
```

### *observation* : By seeing this heatmap there is no correation in between those attributes.

```python
# barplot

sns.set(rc={'figure.figsize':(14,8)})
sns.barplot(data=data, x="Dept", y="salary", hue="certifications")

```

### *observation* :  By seeing above graph the employee working in purchasing department with certification getting more salary.

```python
# boxplot
sns.set(rc={'figure.figsize':(14,8)})
sns.boxplot(data=data, x='Dept', y='salary')
```

```python
#lineplot
sns.set(rc={'figure.figsize':(14,8)})
sns.lineplot(data=data, x="Years_of_ Experience", y="salary",hue="gender")
```

### *observation* : As u can see when years of experience is increasing the employee salary is also increasing.
### In that female are getting more salary as compared to male.
